```python
import umap
import hdbscan
import MDAnalysis as mda
import os
import numpy as np
import matplotlib.pyplot as plt

import sklearn
```

```python
BASE_PATH = r"E:\C_backup\Documents-2\Documents\Simulations\tyloxapol_tx\tyl_7"
SYSTEM_NAME = "100pc_tyl"
```

```python
atom_names = """name C0 C7 C34 C51 C68 C85 C102"""
```

```python
top = os.path.join(BASE_PATH, SYSTEM_NAME, "centered.gro")
trj = os.path.join(BASE_PATH, SYSTEM_NAME, "centered.xtc")
```

```python
u = mda.Universe(top, trj)
```

```python
ag = u.select_atoms(atom_names)
```

```python
ag.positions
```

```python
positions = [ag.positions]

for _ in u.trajectory:
    positions.append(ag.positions)
```

```python
positions = np.array(positions)
```

```python
p = positions.flatten()
```

```python
p
```

```python
reducer = umap.UMAP(n_neighbors=16, min_dist=0.6)

embedded = reducer.fit_transform(p)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(embedded[:, 0], embedded[:, 1], s=0.5)
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=86, cluster_selection_method='leaf')
clusterer.fit(embedded)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(embedded[:, 0], embedded[:, 1], c=clusterer.labels_, cmap='tab10', s=0.5)
```

```python
clusterer.labels_.max()
```

```python
kmeans = sklearn.cluster.KMeans()
```

```python
kmeans_fit = kmeans.fit(embedded)
```

```python
plt.scatter(embedded[:, 0], embedded[:, 1], c=kmeans_fit.labels_, s=0.5)
```
