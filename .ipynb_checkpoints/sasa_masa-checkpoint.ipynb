```python

data = []
for s in systems:
    data.append(np.array(np.load(f'sasa_{surf}_{s}.npy', allow_pickle=True)))

plt.figure(figsize=(10, 6))
for i, d in enumerate(data):
    if i == 0:
        plt.plot(d, label=f'{systems[i]}% TYL7', linewidth=2)
    # elif i == 1:
    #     plt.plot(d, label=f'{tyl_pc}% TYL3 - {tx_pc}% TX100', linewidth=2)
    else:
        tyl_pc, tx_pc = systems[i].split('_')
        plt.plot(d, label=f'{tyl_pc}% TYL7 - {tx_pc}% TX100', linewidth=2)
#     plt.hist(d[420:], rwidth=.9, alpha=.9)
plt.legend()
plt.xlim(0, 600)
plt.ylim(40000, 70000)
plt.legend()
plt.xlabel('Time (ns)')
plt.ylabel('SASA ($\AA^2$)')
plt.savefig(f'{BASE_PATH}/{surf}_sasa.png')
plt.show()
```

```python
colors = ['blue', 'orange', 'green', 'red']
fig, ax = plt.subplots(1, 4, figsize=(12, 3))
for i, d in enumerate(data):
    t = systems[i].split('_')
    title = f'{t[0]} TYL7 - {t[1]} TX0' if i != 0 else f'{t[0]} TYL7'
    ax[i].hist(d, color=colors[i], rwidth=.9)
    ax[i].set_xlim(d.min(), d.max())
    ax[i].set_title(title)
    if i != 0:
        ax[i].set_ylim()
plt.savefig('tyl7_sasa.png')
plt.show()
```
